```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
x
```

```python
x_sq=x**2
x_sq
```

```python
plt.plot(x,x_sq)
plt.show()
```

```python
plt.plot(x,x_sq,marker='o',label='y=x^2')
plt.plot(x,x,marker='o',label='y=x')
plt.xlabel('x')
plt.ylabel('Graph')
plt.title('Comparison of func')
#plt.xlim()=to set limits
plt.legend()
plt.show()
```

```python
plt.plot(x,x,marker='o',c='cyan')
plt.show()
```

```python
plt.scatter(x,x_sq,marker='o',label='y=x^2')
plt.scatter(x,x,marker='o',label='y=x')
plt.xlabel('x')
plt.ylabel('Graph')
plt.title('Comparison of func')
#plt.xlim()=to set limits
plt.legend()
plt.show()
```
